```python
 # show How many COVID-19 vaccine doses were administered in the previous 12 months? 
# That is the first does of vaccine administered In each country 

# method for looping through the ocuntries. 

# first create an emty list to hold country name []
# itterate through the columns to populate the lists 
# create for loop through the list

# answer in the film data base lesson activity 6/10

# results.'collecttionname'

# Connect to your MongoDB database
client = MongoClient("localhost", 27017)

db = client["coviddatadb"] 

collection = db["cleanedcoviddata"]  


# Calculate the date range for the previous 12 months

end_date = dt.date(2023, 8 ,2)
start_date = end_date - timedelta(days=365)
end_date = end_date.strftime('%d/%m/%Y')
start_date = start_date.strftime('%d/%m/%Y')

first_dose_vaccinations = []

# Create the aggregation pipeline
pipeline = [
    {
        "$match": {
            "date": {"$gte": start_date, "$lte": end_date},
            "people_vaccinated": {"$exists": True}  # Filter out entries without vaccination data
        }
    },
    {
        "$group": {
            "_id": "$country",
            "people_vaccinated": {"$sum": "$people_vaccinated"}
        }
    }
]

# Execute the aggregation pipeline
result = list(collection.aggregate(pipeline))

# Print the total people vaccinated for each country
for entry in result:
    country = entry["_id"]
    total_vaccinated_dose = entry["people_vaccinated"]
    print(f"Country: {country}, Total People Vaccinated with the first does: {total_vaccinated_dose}")
    
    first_dose_vaccinations.append ({"country": country , " first dose vaccinations ": total_vaccinated_dose})
    

```

# Relationship between GDP and COVID-19 Cases. 


Which countries has the least cases of covid-19?
Which countries has the most cases of covid-19 ?
Display a graph showing Relationship between GDP and COVID-19 Cases. 
Columns needed: total_cases_per_million, gdp_per_capita


```python
# Which countries has the least cases of covid-19? 

# Create the aggregation pipeline

pipeline_least = [
    {
        "$match": {
            "total_cases": {"$gt": 0},  # Exclude entries with zero cases
            "total_cases_per_million": {"$exists": True} 
        }
    },
    {
        "$sort": {"total_cases_per_million": 1}
    },
    {
        "$limit": 1
    }
]

# Execute the aggregation pipeline
result_least = list(collection.aggregate(pipeline_least))

# Print the country with the least cases
print("Country with the Least Cases of COVID-19:")
if result_least:
    print(result_least[0]["country"])
else:
    print("No valid data found.")
```

```python
# Which countries has the most cases of covid-19 ?
# Create the aggregation pipeline
pipeline_most = [
    {
        "$match": {
            "total_cases_per_million": {"$exists": True}  # Filter out entries without case data
        }
    },
    {
        "$sort": {"total_cases_per_million": -1}
    },
    {
        "$limit": 1
    }
]

# Execute the aggregation pipeline
result_most = list(collection.aggregate(pipeline_most))

# Print the country with the most cases
print("Country with the Most Cases of COVID-19:")
if result_most:
    print(result_most[0]["country"])
else:
    print("No valid data found.")
```

Where average GDP increases as average COVID-19 cases per million also increase, suggests a positive correlation between these two variables. In other words, countries with higher average GDP tend to have higher average COVID-19 cases per million.

The fact that the scatter plot is more dense in the range of 0-20000 GDP indicates that many countries with lower GDP values also have relatively higher COVID-19 cases. This could be due to various factors such as population density, healthcare infrastructure, public health measures, and economic activities. It's important to note that correlation does not imply causation, so while there appears to be a relationship, it doesn't necessarily mean that higher GDP directly causes higher COVID-19 cases.

Further analysis and consideration of other variables might be needed to understand the underlying reasons for this pattern. It's also worth looking into potential outliers or specific cases that deviate from the overall trend to gain a more nuanced understanding.

```python
# # Query the collection to retrieve the required data
# Create the aggregation pipeline for the highest stringency index
pipeline_highest = [
    {
        "$match": {
            "stringency_index": {"$exists": True}
        }
    },
    {
        "$sort": {"stringency_index": -1}
    },
    {
        "$limit": 1
    }
]

# Execute the aggregation pipeline for the highest stringency index
result_highest = list(collection.aggregate(pipeline_highest))

# Print the country with the highest stringency index
print("Country with the Highest Stringency Index:")
if result_highest:
    print(result_highest[0]["country"])
else:
    print("No valid data found.")

```

```python
# Create the aggregation pipeline for the lowest stringency index
pipeline_lowest = [
    {
        "$match": {
            "stringency_index": {"$exists": True}
        }
    },
    {
        "$sort": {"stringency_index": 1}
    },
    {
        "$limit": 1
    }
]

# Execute the aggregation pipeline for the lowest stringency index
result_lowest = list(collection.aggregate(pipeline_lowest))

# Print the country with the lowest stringency index
print("Country with the Lowest Stringency Index:")
if result_lowest:
    print(result_lowest[0]["country"])
else:
    print("No valid data found.")
```
