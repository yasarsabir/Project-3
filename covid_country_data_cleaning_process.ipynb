```python
import pandas as pd 
```

```python
#creating the paths to the covid and country files

world_country_path = 'world_country_lat_long.csv'
covid_data_path = 'owid-covid-data-filtered.csv'
```

```python
#reading the country data into pandas dataframe
world_country_df = pd.read_csv('world_country_lat_long.csv')

world_country_df.head()
```

```python
# selecting the country,latitude and longitude columns

country_df = world_country_df[['country','latitude','longitude']]

country_df
```

```python
#checking for missing values

missing_values = country_df.isnull().sum()

missing_values
```

```python
#removing the row with no latitude and longitude data

country_df = country_df[country_df['country'] != 'U.S. Minor Outlying Islands']
```

```python
#verifying missing data

missing_values = country_df.isnull().sum()

missing_values
```

```python
#reading the covid data into pandas dataframe

covid_df = pd.read_csv(covid_data_path)

covid_df.head()
```

```python
#renaming the covid data column to ensure that the name tallies with the country data column 
covid_df= covid_df.rename(columns = {'location':'country'})

covid_df.head()
```

```python
merged_df = pd.merge(covid_df, country_df, on= 'country', how= 'left')

merged_df.head()
```
